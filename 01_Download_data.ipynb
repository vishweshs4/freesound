Let's see what the sampling rates and durations of the other files in the dataset are.

```python
from pathlib import Path
```

```python
%%time

trn_srs = []
trn_lengths = []

for path in Path('data/audio_train').iterdir():
    x, sr = librosa.core.load(path, sr=None)
    trn_srs.append(sr)
    trn_lengths.append(x.shape[0] / sr)
    
tst_srs = []
tst_lengths = []

for path in Path('data/audio_test').iterdir():
    x, sr = librosa.core.load(path, sr=None)
    tst_srs.append(sr)
    tst_lengths.append(x.shape[0] / sr)
```

```python
import numpy as np
```

```python
np.all(np.array(trn_srs) == 44100), np.all(np.array(tst_srs) == 44100)
```

The sampling rates of all the files are the same, 44100.

Working on subsequent notebooks it turned out that there are three files in the test set that are corrupt. We could address this down the road, at submission time creation, but it is best to address this right now, if we can. This way we never have to worry about this again.

Less congnitive load, not having to maintain state (name of the files), making the code simpler down the road - all are a clear win to me. Sure, this is maybe a very simple scenario, one that is not too much of a hassle. Nonetheless, the reasoning stands.

Let's subsitute the corrupted files with something that we will be able to load to predict on.

```python
for fn in ['0b0427e2.wav', '6ea0099f.wav', 'b39975f5.wav']:
    x = np.zeros(2 * sr)
    librosa.output.write_wav(f'data/audio_test/{fn}', x, sr, norm=True)
```
